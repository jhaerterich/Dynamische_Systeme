```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

# Define the function
def f(x):
    if x == 0:
        return 1
    else:
        return np.sin(x) / x


# Define the x-axis values
x = np.linspace(0.01, 50, 2000)

# Evaluate the function for the x-axis values
y = [f(xi) for xi in x]

# Create the plot
fig, ax = plt.subplots()
ax.plot(x, y, 'r', linewidth=2)
ax.fill_between(x, y, 0, alpha=0.2)
ax.set_xlim([0, 50])
ax.set_ylim([-0.3, 1.1])
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Uneigentliches Integral von f(x)=sin(x)/x')
```
